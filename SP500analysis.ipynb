```python
import scipy as sc
from scipy.optimize import minimize

def MarkPortOpt(ret,cov,r,m, silent = False):
    def EReturn(w):
        EReturn = ret @ w.T
        return EReturn

# Function to compute the portfolio standard deviation
    def PVol(w):
        pvar = w.T @ cov @ w
        return np.sqrt(pvar)
    # Constraints
    def constraint1(w):
        return np.sum(w) - 1.0 # budget constraint   
    def constraint2(w):
        return 1.0 - np.sum(w) # budget constraint
    def constraint3(w):
        return w[:-1] # nonnegative constraint placed on stock returns only
    def constraint4(w):
        diff = EReturn(w) - r
        return diff # return constraint 
    def constraint5(w):
        return m - w


    con1 = {'type': 'ineq', 'fun': constraint1}
    con2 = {'type': 'ineq', 'fun': constraint2}
    con3 = {'type': 'ineq', 'fun': constraint3}
    con4 = {'type': 'ineq', 'fun': constraint4}
    con5 = {'type':'ineq','fun':constraint5}
    cons = ([con1,con2,con3,con4,con5])

    # Initial x0
    w0 = np.ones((1,len(ret)))

    # Solve the problem
    sol = minimize(PVol, w0, method='SLSQP',constraints=cons)
    
    # Whether the solution will be printed
    if(not silent):
        print("Solution to the Markowitz Problem with r =  ", round(r*100,3), "%:")
        print(sol)
        print("")
    elif (not sol['success']): #check if the optimizer exist successfully
        print("WARNING:  the optimizer did NOT exit successfully!!")
        
    return sol
```

```python
def MaxSR(ret,cov, rf , silent = False):

    #Objective Function
    def PVol(w):
        pvar = w.T @ cov @ w
        return np.sqrt(pvar)

    def SR(w):
        excess_ret = ret @ w.T - rf 
        SR = (excess_ret.mean())/(PVol(w))
        return -SR
    
    n = len(ret)    
    #Bounds
    bnds = ((0,1),)*n # nonnegativity constraint

    # Constraints
    def constraint1(w):
        return np.sum(w) - 1.0 # budget constraint   
    cons = {'type': 'eq', 'fun': constraint1}
    
    # Initial x0
    w0 = np.ones((1,len(ret)))
    
    # Solve the problem
    sol = minimize(SR, w0, method='SLSQP', bounds=bnds, constraints=cons)
    
    # Whether the solution will be printed
    if(not silent):
        print("Solution to the Max Sharpe Ratio Problem is:")
        print(sol)
        print("")
    elif (not sol['success']): #check if the optimizer exist successfully
        print("WARNING:  the optimizer did NOT exit successfully!!")
        
    return sol
```
